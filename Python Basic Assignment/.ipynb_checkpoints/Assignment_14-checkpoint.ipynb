# What does RGBA stand for?


'RGBA' is abbreviation for 'red,green,blue,alpha(transparency of the color)'. 

In python it corresponds to the tuple of 4 values ranging from '0 to 255' 

# From the Pillow module, how do you get the RGBA value of any images?


```python
from PIL import Image, ImageDraw, ImageFilter

imgobj = Image.open('C:\Piyush\Personal\Study\AIML\Krish Naik\IneuronAssignments\Python Basic Assignment\Files\Source\lena.jpg')
pixels = imgobj.convert('RGBA')
data = imgobj.getdata()
lofpixels = []
for pixel in data:
    lofpixels.extend(pixel)
    
print(lofpixels)
```

# What is a box tuple, and how does it work?



A box tuple is a tuple value of four integers: the left edge x-coordinate, the top edge y-coordinate, the width, and the height, respectively

# What module contains Pillows shape-drawing code?


The ImageDraw module contains code to draw on images.

```python
import sys
from PIL import Image, ImageDraw

with Image.open('C:\Piyush\Personal\Study\AIML\Krish Naik\IneuronAssignments\Python Basic Assignment\Files\Source\lena.jpg') as im:

    draw = ImageDraw.Draw(im)
    draw.line((0, 0) + im.size, fill=128)
    draw.line((0, im.size[1], im.size[0], 0), fill=128)

    # write to stdout
    im.save('C:\Piyush\Personal\Study\AIML\Krish Naik\IneuronAssignments\Python Basic Assignment\Files\Source\lena_modified.jpg')
```

# Image objects do not have drawing methods. What kind of object does? How do you get this kind of object?

ImageDraw objects have shape-drawing methods such as point(), line(), or rectangle(). 

They are returned by passing the Image object to the 'ImageDraw.Draw()' function.
